```python
from PIL import Image, ImageEnhance
import matplotlib.pyplot as plt
import numpy as np

# Function to adjust brightness
def adjust_brightness(image, factor):
    """Adjust brightness by the given factor."""
    enhancer = ImageEnhance.Brightness(image)
    print(f"Enhancer object created: {enhancer}")

    # Convert the image to an array for math inspection
    image_array = np.array(image)
    print(f"Original image array (first 5 pixels): \n{image_array[:5]}")

    # Apply brightness enhancement
    bright_image = enhancer.enhance(factor)

    # Convert the enhanced image to an array for math inspection
    bright_image_array = np.array(bright_image)
    print(f"Enhanced image array (first 5 pixels): \n{bright_image_array[:5]}")

    return bright_image

# Open an image (replace 'your_image.jpg' with your image path)
image_path = 'test_img.png'
print(f"Opening image from: {image_path}")
image = Image.open(image_path)
print(f"Image opened: {image}")
print(f"Image size: {image.size}, Image mode: {image.mode}")

# Convert image to numpy array for printing values
image_array = np.array(image)
print(f"Original image array (first 5 pixels): \n{image_array[:5]}")

# Adjust brightness by a factor of 5.0 (e.g., make it brighter)
brightness_factor = 5.0
print(f"Adjusting brightness with factor: {brightness_factor}")
bright_image = adjust_brightness(image, brightness_factor)

# Show the original and adjusted images
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
print(f"Figure object: {fig}")
print(f"Axes object: {axes}")

# Show original image
axes[0].imshow(image)
axes[0].set_title("Original Image")
axes[0].axis('off')
print("Displaying original image")

# Show adjusted image
axes[1].imshow(bright_image)
axes[1].set_title("Adjusted Image")
axes[1].axis('off')
print("Displaying adjusted image")

# Show the plot
plt.show()
print("Displaying plot")

```
